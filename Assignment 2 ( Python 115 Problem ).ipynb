# 18. Quadratic Equation Solver: Write a Python program that takes the coefficients (a, b, c) of a quadratic equation as input and calculates and prints the real roots (if they exist) or a message indicating the complex roots.

```python
import math

a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))

print("The quadratic equation is: {}x² + {}x + {} = 0".format(a, b, c))

if a == 0:
    print("This is not a quadratic equation.")
else:
    discriminant = b**2 - 4*a*c

    if discriminant >= 0:
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        if discriminant == 0:
            print("The equation has one real root:", root1)
        else:
            print("The equation has two real roots:", root1, "and", root2)
    else:
        print("The equation has complex roots. No real roots exist.")
```

# 24. Fibonacci Sequence: Write a Python program using a for loop to generate the Fibonacci sequence up to a given limit N.

```python
N = int(input("Enter the limit N for the Fibonacci sequence: "))

fib_sequence = [0, 1]

if N == 0:
    fib_sequence = [0]
elif N == 1:
    fib_sequence = [0, 1]

for i in range(2, N):
    next_term = fib_sequence[i-1] + fib_sequence[i-2]
    if next_term > N :
        break
    else:
        fib_sequence.append(next_term)

print("Fibonacci sequence up to", N, "terms is:")
print(fib_sequence)
```

```python
n = int(input("Enter how many numbers you want in the list: "))


numbers = []

for i in range(n):
    num = int(input(f"Enter number {i+1}: "))
    numbers.append(num)

print("\nThe list of numbers is:", numbers)

# Initialize variables
total = 0
maximum = numbers[0]
minimum = numbers[0]

# Calculate sum, max, and min using a for loop
for num in numbers:
    total += num
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

average = total / n

# Display results
print("Sum of the numbers:", total)
print("Average of the numbers:", average)
print("Maximum value:", maximum)
print("Minimum value:", minimum)
```

```python
A = [[1, 2, 3],
     [4, 5, 6]]

B = [[7, 8],
     [9, 10],
     [11, 12]]

result = [[0, 0],
          [0, 0]]

for i in range(len(A)):
    for j in range(len(B[0])):
        for k in range(len(B)):
            result[i][j] += A[i][k] * B[k][j]

for row in result:
    print(row)
```

# 46. List Element Frequency: Given a nested list containing lists of integers, write a Python program to count the frequency of each element in the entire nested list.

```python
nested_list = [[1, 2, 3], [2, 3, 4], [1, 4, 4, 5]]
print("Nested list:", nested_list)

frequency = {}

for sublist in nested_list:
    for item in sublist:
        if item in frequency:
            frequency[item] += 1
        else:
            frequency[item] = 1

print("Element frequencies:")
for key, value in frequency.items():
    print(f"{key}: {value}")
```

# 47. Transpose Matrix: Write a Python program to transpose a given matrix represented as a nested list.



```python
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

matrix = []
print("Enter elements of the matrix:")
for i in range(rows):
    row = []
    for j in range(cols):
        row.append(int(input(f"Element [{i+1},{j+1}]: ")))
    matrix.append(row)

print("\nOriginal matrix:")
for row in matrix:
    print(row)

# Transpose the matrix
transpose = []
for j in range(cols):
    new_row = []
    for i in range(rows):
        new_row.append(matrix[i][j])
    transpose.append(new_row)

print("\nTransposed matrix:")
for row in transpose:
    print(row)
```

# 108. Math Operations: Write a Python function called `math_operations` that takes three numbers and a string representing an operation (‘add’, ‘subtract’, ‘multiply’, or ‘divide’). The function should return the result of the specified operation on the three numbers. Implement the math operations as nested functions

```python
def math_operations(a, b, c, operation):
    def add(x, y, z):
        return x + y + z

    def subtract(x, y, z):
        return x - y - z

    def multiply(x, y, z):
        return x * y * z

    def divide(x, y, z):
        if y == 0 or z == 0:
            return "Error: Division by zero!"
        return x / y / z

    if operation == "add":
        return add(a, b, c)
    elif operation == "subtract":
        return subtract(a, b, c)
    elif operation == "multiply":
        return multiply(a, b, c)
    elif operation == "divide":
        return divide(a, b, c)
    else:
        return "Invalid operation!"

while True:
    user_input = input("Enter three numbers and an operation (add, subtract, multiply, divide) or type 'stop' to end: ")
    
    if user_input.lower() == "stop":
        print("Program stopped.")
        break

    try:
        parts = user_input.split()
        if len(parts) != 4:
            print("Please enter exactly three numbers followed by an operation.")
            continue
        
        a, b, c = map(float, parts[:3])
        operation = parts[3].lower()

        print(f"Given numbers: {a}, {b}, {c}")
        print(f"Operation: {operation}")
        print(f"Result: {math_operations(a, b, c, operation)}")
    except ValueError:
        print("Invalid input! Please enter numbers followed by a valid operation.")
```

# 110. Temperature Converter: Write a Python function called `temperature_converter` that takes a temperature value and a string representing the scale (‘C’ for Celsius or ‘F’ for Fahrenheit) as input. The function should convert the temperature from one scale to the other using nested functions and return the converted value.

```python
def temperature_converter(temp, scale):
    def to_celsius(f):
        return (f - 32) * 5 / 9

    def to_fahrenheit(c):
        return (c * 9 / 5) + 32

    if scale.upper() == "C":
        return to_fahrenheit(temp)
    elif scale.upper() == "F":
        return to_celsius(temp)
    else:
        return "Invalid scale! Use 'C' for Celsius or 'F' for Fahrenheit."

while True:
    user_input = input("Enter temperature and scale (e.g., 100 C or 212 F) or type 'stop' to end: ")
    
    if user_input.lower() == "stop":
        print("Program stopped.")
        break

    try:
        parts = user_input.split()
        if len(parts) != 2:
            print("Invalid input! Enter a number followed by a scale (C or F).")
            continue
        
        temp = float(parts[0])
        scale = parts[1].upper()

        print(f"Given input: {temp} {scale}")
        converted = temperature_converter(temp, scale)
        if scale == "C":
            print(f"{temp}°C is {converted:.2f}°F")
        elif scale == "F":
            print(f"{temp}°F is {converted:.2f}°C")
        else:
            print(converted)
    except ValueError:
        print("Invalid input! Please enter a valid number and scale.")
```

# 111. Factorial Calculation: Write a recursive Python function called `factorial` that takes a non-negative integer as input and returns its factorial

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

while True:
    user_input = input("Enter a non-negative integer to find its factorial (or type 'stop' to end): ")
    
    if user_input.lower() == "stop":
        print("Program stopped.")
        break
    
    if user_input.isdigit():
        number = int(user_input)
        print(f"Given input: {number}")
        print(f"Factorial of {number} is: {factorial(number)}")
    else:
        print("Invalid input! Please enter a non-negative integer or 'stop'.")
```

# 112. Fibonacci Series: Write a recursive Python function called `Fibonacci` that takes an integer N as input and returns the Nth number in the Fibonacci series. The Fibonacci series is defined as follows: F(0) = 0, F(1) = 1, and F(n) = F(n-1) + F(n-2) for n > 1.

```python
def Fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return Fibonacci(n - 1) + Fibonacci(n - 2)

while True:
    user_input = input("Enter a non-negative integer N to find the Nth Fibonacci number (or type 'stop' to end): ")
    
    if user_input.lower() == "stop":
        print("Program stopped.")
        break
    
    if user_input.isdigit():
        N = int(user_input)
        print(f"Given input: {N}")
        print(f"The {N}th Fibonacci number is: {Fibonacci(N)}")
    else:
        print("Invalid input! Please enter a non-negative integer or 'stop'.")
```

# 113. The sum of Digits: Write a recursive Python function called `sum_of_digits` that takes an integer as input and returns the sum of its digits.

```python
def sum_of_digits(n):
    n = abs(n)
    if n == 0:
        return 0
    else:
        return n % 10 + sum_of_digits(n // 10)

while True:
    user_input = input("Enter an integer to find the sum of its digits (or type 'stop' to end): ")
    
    if user_input.lower() == "stop":
        print("Program stopped.")
        break
    
    try:
        number = int(user_input)
        print(f"Given input: {number}")
        print(f"Sum of digits of {number} is: {sum_of_digits(number)}")
    except ValueError:
        print("Invalid input! Please enter an integer or 'stop'.")
```

# 114. Binary Search: Write a recursive Python function called `binary_search` that takes a sorted list and a target value as input and returns the index of the target value in the list using binary search. If the target value is not in the list, return -1.

```python
def binary_search(arr, target, low=0, high=None):
    if high is None:
        high = len(arr) - 1

    if low > high:
        return -1

    mid = (low + high) // 2

    if arr[mid] == target:
        return mid
    elif arr[mid] > target:
        return binary_search(arr, target, low, mid - 1)
    else:
        return binary_search(arr, target, mid + 1, high)

while True:
    user_input = input("Enter a sorted list of integers separated by spaces (or type 'stop' to end): ")
    
    if user_input.lower() == "stop":
        print("Program stopped.")
        break

    try:
        arr = [int(x) for x in user_input.split()]
        target_input = input("Enter the target value: ")
        target = int(target_input)
        print(f"Given list: {arr}")
        print(f"Target value: {target}")
        index = binary_search(arr, target)
        if index != -1:
            print(f"Target found at index: {index}")
        else:
            print("Target not found in the list.")
    except ValueError:
        print("Invalid input! Please enter integers only.")
```

# 115. Power Calculation: Write a recursive Python function called `power` that takes two positive integers, base and exponent, as input and returns the value of base raised to the exponent.

```python
def power(base, exponent):
    if exponent == 0:
        return 1
    else:
        return base * power(base, exponent - 1)

while True:
    user_input = input("Enter base and exponent separated by space (or type 'stop' to end): ")
    
    if user_input.lower() == "stop":
        print("Program stopped.")
        break

    try:
        base, exponent = map(int, user_input.split())
        if base < 0 or exponent < 0:
            print("Please enter positive integers only.")
            continue
        print(f"Given input: base = {base}, exponent = {exponent}")
        print(f"{base}^{exponent} = {power(base, exponent)}")
    except ValueError:
        print("Invalid input! Please enter two positive integers separated by space.")
```
